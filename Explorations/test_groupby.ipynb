```python
import pandas as pd
test = pd.DataFrame(dict(country=["sweden", "sweden", "sweden", "iceland", "iceland", "iceland", "norway", "norway", "norway"], positivity_rate=[3, 4, 5, 6, 7, 8, 9, 10, 11], new_cases=[5, 3, 5, 6, 3, 4, 5, 6, 1], population=[176, 176, 176, 245, 245, 245, 567, 567, 567]))
test
```

```python
test["cases_as_percentage_of_population"] = test["new_cases"]/test["population"]
test
```

```python
test2 = pd.DataFrame(test["positivity_rate"].groupby(test["country"]).mean())
test2["total_cases_in_percent"] = test["cases_as_percentage_of_population"].groupby(test["country"]).sum()*100
test2.reset_index()
```
