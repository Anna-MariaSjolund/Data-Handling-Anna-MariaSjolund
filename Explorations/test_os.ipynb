```python
test_gender = pd.DataFrame({"ID":[1, 1, 2, 3, 4], "Name":["Anna", "Anna", "Anna", "Maria", "Patrik"], "Sex":["F", "F", "F", "F", "M"]})
test_gender
```

```python
gender = ["Female", "Male"]
count = [len(test_gender[test_gender["Sex"] == "F"].groupby(test_gender["ID"])), 
        len(test_gender[test_gender["Sex"] == "M"].groupby(test_gender["ID"]))]

gender_count_ID = pd.DataFrame(dict(Gender=gender, Count=count))
gender_count_ID
```
